## analysis  teamwise

```python
sorted(matches['season'].unique())
```

```python
season_wins = matches.groupby(['season', 'winner'])['winner'].count().reset_index(name='wins')

# Find the team with the highest wins in each season
season_winner = season_wins.loc[season_wins.groupby('season')['wins'].idxmax()]
print(season_winner)
```

```python
match_winner=matches.groupby('winner')['winner'].count().nlargest(5)
match_winner_df=match_winner.reset_index(name='count')
plt.figure(figsize=(8, 6))
for i, value in enumerate(match_winner_df['count']):
    plt.text(i, value, str(value), ha='center', va='bottom')
    
plt.bar(match_winner_df['winner'], match_winner_df['count'])
plt.title('most winning of the match')
plt.xlabel('winning team')
plt.ylabel('total time')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
grouped = delivers.groupby('batting_team')['total_runs'].sum().nlargest(5)
grouped_df = grouped.reset_index(name='count')
plt.figure(figsize=(8, 6))
for i, value in enumerate(grouped_df['count']):
    plt.text(i, value, str(value), ha='center', va='bottom')
    
plt.bar(grouped_df['batting_team'], grouped_df['count'])
plt.title('highest total runs')
plt.xlabel('batting_team')
plt.ylabel('total run')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
bowling = delivers.groupby('bowling_team')['player_dismissed'].count().nlargest(5)
bowling_df = bowling.reset_index(name='count')
plt.figure(figsize=(8, 6))
for i, value in enumerate(bowling_df['count']):
    plt.text(i, value, str(value), ha='center', va='bottom')
    
plt.bar(bowling_df['bowling_team'], bowling_df['count'])
plt.title('highest wicket taking team')
plt.xlabel('bowling team')
plt.ylabel('total wickets')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
finals = matches.drop_duplicates(subset=['season'],keep='last')
finals = finals[['id','season','team1','team2','toss_winner','toss_decision','winner']]
most_finals = pd.concat([finals['team1'],finals['team2']]).value_counts().reset_index()
most_finals.rename({'index':'teams',0:'final played'},axis=1,inplace=True)
wins=finals['winner'].value_counts().reset_index()
wins.rename({'index':'teams'},axis=1,inplace=True)
most_finals=most_finals.merge(wins,left_on='teams',right_on='teams',how='inner')
most_finals
```

```python
bar_width = 0.35
fig, ax = plt.subplots(figsize=(10, 6))

x = np.arange(len(most_finals['teams']))
bar1=ax.bar(x - bar_width/2, most_finals['final played'], bar_width, label='final played')
bar2=ax.bar(x + bar_width/2, most_finals['winner'], bar_width, label='winner')
for bar in bar1 + bar2:
    height = bar.get_height()
    ax.annotate(f'{height}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')
ax.set_xlabel('teams')
ax.set_ylabel('no. of times')
ax.set_title('final winning ratio')
ax.set_xticks(x)
ax.set_xticklabels(most_finals['teams'])
ax.legend()

plt.show()
```

```python
matche_plays = pd.concat([matches['team1'],matches['team2']]).value_counts().reset_index()
matche_plays.rename({'index':'teams',0:'no. of match played'},axis=1,inplace=True)
win=matches['winner'].value_counts().reset_index()
win.rename({'index':'teams'},axis=1,inplace=True)
most_wins=matche_plays.merge(win,left_on='teams',right_on='teams',how='inner')
most_wins['win percentage']=(most_wins['winner']*100)/most_wins['no. of match played']
most_wins=most_wins.sort_values('win percentage',ascending=False)
most_wins
```

```python
grouped = delivers.groupby('batting_team')['total_runs'].sum()
grouped_df = grouped.reset_index(name='count')
merged_df=matche_plays.merge(grouped_df,left_on='teams',right_on='batting_team',how='inner')
merged_df['average_run']=merged_df['count']//merged_df['no. of match played']
merged_df=merged_df.sort_values('average_run',ascending=False)
merged_df
```

```python
def team1_vs_team2_team_battle(team1, team2):
    filtered_df = matches[((matches['team1'] == team1) & (matches['team2'] == team2)) | ((matches['team1'] == team2) & (matches['team2'] == team1))]
    filtered_df = filtered_df[['id', 'season', 'team1', 'team2', 'toss_winner', 'toss_decision', 'winner']]
    matche_plays = pd.concat([filtered_df['team1'],filtered_df['team2']]).value_counts().reset_index()
    matche_plays.rename({'index':'teams',0:'no. of match played'},axis=1,inplace=True)
    win=filtered_df['winner'].value_counts().reset_index()
    win.rename({'index':'teams'},axis=1,inplace=True)
    most_wins=matche_plays.merge(win,left_on='teams',right_on='teams',how='inner')
    most_wins['win percentage']=(most_wins['winner']*100)/most_wins['no. of match played']
    most_wins=most_wins.sort_values('win percentage',ascending=False)
    print(most_wins)
    sns.set(style="whitegrid")  
    plt.figure(figsize=(4, 4))  
    plt.pie(most_wins['win percentage'], labels=most_wins['winner'], autopct='%1.1f%%', startangle=140)
    plt.title("match winning percentage")
    plt.axis('equal')  
    plt.show()

    
if __name__ == "__main__":    
    
    a=str(input('Enter team1 name = '))
    b=str(input('Enter team2 name = '))
    team1_vs_team2_team_battle(a,b)
```
